### Create hierarchy of occupations based on rules ###

```python
titles=df['2021 National Employment Matrix title']
codes=df['2021 National Employment Matrix code']
occupations=df['Occupation type']
p=-1
c=0

dict1={}
for title in titles:
    
# all occupations summary
    if c==0:
        
        dict1['title']=titles[c]
        dict1['code']=codes[c]
        dict1['occupation']=occupations[c]
        dict1['layer']='layer_1'
        dict1['children']=[]
        
    else:
        c_code=codes[c]
        c_title=titles[c]
        c_occupation=occupations[c]



        previous_code=codes[p]

        # print(f"Current Code: {c_code}")

        current_first_two=c_code.split("-")[0]
        # print(f"Current first two: {current_first_two}")

        current_last_four=c_code.split("-")[1]
        # print(f"Current last four: {current_last_four}")

        # print(f"Previous Code: {previous_code}")

        previous_first_two=previous_code.split("-")[0]
        # print(f"Previous first two: {previous_first_two}")

        d=0

        third_digit=current_last_four[0]
        last_two_digits=f"{current_last_four[2]}{current_last_four[3]}"
        # print (f"This is the last two digits {last_two_digits}")

        middle_digits=f"{current_last_four[0]}{current_last_four[1]}"
        last_digit=current_last_four[3]

        th_fo_fi_digits= f"{current_last_four[0]}{current_last_four[1]}{current_last_four[2]}"


    #         logic step 1
        if current_last_four == "0000":

            dict2={}
            dict2['title']=c_title
            dict2['code']=c_code
            dict2['occupation']=c_occupation
            dict2['layer']="layer2"
            dict2['group']=current_first_two
            dict2['children']=[]

            dict1['children'].append(dict2)

    #         logic step 2
        elif third_digit != 0 and last_two_digits =="00":

            dict3={}
            dict3['title']=c_title
            dict3['code']=c_code
            dict3['occupation']=c_occupation
            dict3['layer']="layer3"
            dict3['group']=current_first_two
            dict3['children']=[]
            dict3['m_digits']=middle_digits


            for child in dict1['children']:
                # print(child)
                if current_first_two == child['group']:
                    child['children'].append(dict3)
    #         logic step 3
    #         elif 'children' in dict1:
        else:
            for child in dict1['children']:

                for child2 in child['children']:

                    child2_m_digits=child2['m_digits']


                    if middle_digits== child2_m_digits:
                    # logic step 3A


                        if last_digit==0 and c_occupation==['Summary']:
                            dict4={}
                            dict4['title']=c_title
                            dict4['code']=c_code
                            dict4['occupation']=c_occupation
                            dict4['layer']="layer4"
                            dict4['group']=current_first_two
                            dict4['children']=[]
                            dict4['m_digits']=middle_digits
                            dict4['last_digit']=last_digit

                            child2['children'].append(dict4)


                        elif last_digit == 0 and c_occupation==['Line item']:
                            dict4={}
                            dict4['title']=c_title
                            dict4['code']=c_code
                            dict4['occupation']=c_occupation
                            dict4['layer']="layer4"
                            dict4['group']=current_first_two
                            dict4['children']=[]
                            dict4['m_digits']=middle_digits
                            dict4['last_digit']=last_digit

                            child2['children'].append(dict4)

                        elif last_digit != 0 :
                            dict4={}
                            dict4['title']=c_title
                            dict4['code']=c_code
                            dict4['occupation']=c_occupation
                            dict4['layer']="layer4"
                            dict4['group']=current_first_two
                            dict4['children']=[]
                            dict4['m_digits']=middle_digits
                            dict4['last_digit']=last_digit

                            child2['children'].append(dict4)

                        # if 'children' in child2 and middle_digits== child2_m_digits  and current_first_two == child2['group'] and child2['code']!=c_code:

                        for child3 in child2['children']:


                            child_last_four=child3['code'].split("-")[1]
                            child_thfofi=f"{child_last_four[0]}{child_last_four[1]}{child_last_four[2]}"
                            child_last_one=f"{child_last_four[3]}"
                            if child_thfofi==th_fo_fi_digits and 'children' in child3  and current_first_two == child3['group'] and child3['code']!=c_code and child_last_one=="0":
                                dict5={}
                                dict5['title']=c_title
                                dict5['code']=c_code
                                dict5['occupation']=c_occupation
                                dict5['layer']="layer5"
                                dict5['group']=current_first_two
                                dict5['children']=[]
                                dict5['m_digits']=middle_digits
                                dict5['last_digit']=last_digit

                                child3['children'].append(dict5)







            # print("-----------")



    c+=1
    p+=1
# dict1
```

### Create occupation mapping hieratchy ###

```python
ls=[]
titles=[]
# if 'children' in df:
# df['children']
for i,row in df.iterrows():
    # print(row['title'])
    # print(row['children']['title'])
    
    top_parent=row['title']
    parent2=row['children']['title']
    
    # print(type(row['children']['children']))
    # print ("----------")
    
    
    for child in range(len(row['children']['children'])):
        
        # print(child)

        # print(row['children']['children'][child]['title'])
        parent3=row['children']['children'][child]['title']
        # print(row['children']['children'][child]['children'])
#         print ("----------")
#         print ("----------")
#         print ("----------")
        for child2 in range(len(row['children']['children'][child]['children'])):
            # print(child2)
            # print(child2, row['children']['children'][child]['children'][child2]['title'])
            parent4=row['children']['children'][child]['children'][child2]['title']
            
            if len(row['children']['children'][child]['children'][child2]['children'])!=0:
                # print(row['children']['children'][child]['children'][child2]['children'])
                
                for child3 in range(len(row['children']['children'][child]['children'][child2]['children'])):
                    # print(f"---- {child3}: {row['children']['children'][child]['children'][child2]['children'][child3]['title']}")
                    parent5=row['children']['children'][child]['children'][child2]['children'][child3]['title']
                    dict={}
                    dict['top_parent']=top_parent
                    dict['second_level']=parent2
                    dict['third_level']=parent3
                    dict['fourth_level']=parent4
                    dict['fifth_level']=parent5
                    
                    ls.append(dict)
            else:
                    dict={}
                    dict['top_parent']=top_parent
                    dict['second_level']=parent2
                    dict['third_level']=parent3
                    dict['fourth_level']=parent4
                    dict['fifth_level']=parent4
                    ls.append(dict)
                
    # print("----------------")
pd.DataFrame(ls)
```
